### Load libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load data from admissions.csv 

```python
!!cd
```

```python
df = pd.read_csv("admission.csv")
```

### Build Machine Learning Model 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Model with all features 

```python
X = df[['Gre','Toefl','Cgpa','Rating','Sop','Lor','Research']]
y = df['Chance'] * 100
```

```python
# Split data into train and test 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
## Fit model or Training Model
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
## get MSE, R2 and MAE
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error : {mse:0.2f}")

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error : {mae:0.2f}")

r2score = r2_score(y_test, y_pred)
print(f"R2 Score: {r2score:0.2f}")
```

## Standardization of DataSet

### Scale value - Gre, Toefl, Cgpa, Rating

```python
from sklearn.preprocessing import scale
```

```python
X = df[['Gre','Toefl','Cgpa']]
y = df['Chance'] * 100
```

```python
sns.distplot(X['Gre'])
```

```python
X['Gre'] = scale(X['Gre'])
X['Toefl'] = scale(X['Toefl'])
X['Cgpa'] = scale(X['Cgpa'])
```

```python
sns.distplot(X['Gre'])
```

```python
sns.distplot(X['Toefl'])
```

```python
# Split data into train and test 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
## Fit model or Training Model
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
score = r2_score(y_test,y_pred)
print(f"R2 Score: {score:0.2f}")
```

```python
mse = mean_squared_error(y_test,y_pred)
print(f"Mean Squared Error : {mse:0.2f}")
```
