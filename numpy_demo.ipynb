```python
def prime(a):
    lst = []
    for n in a:
        for i in range(2,n//2 + 1):
            if n % i == 0 :
                lst.append(False)
                break
        else:
            lst.append(True)
        
    return lst
```

```python
a = np.random.randint(1,100,10)
a [prime(a)]
```
